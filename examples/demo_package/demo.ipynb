## 2. Creating Custom Widgets

Your package can define its own widgets easily:

```python
from syft_widget import APIDisplay, register_endpoint

# Register a custom endpoint
@register_endpoint("/custom/data")
def get_custom_data():
    import random
    return {
        "value": random.randint(1, 100),
        "message": "Hello from custom endpoint!"
    }

# Create a custom widget
class CustomWidget(APIDisplay):
    def __init__(self):
        super().__init__(endpoints=["/custom/data"])
    
    def render_content(self, data, server_type="checkpoint"):
        custom = data.get("/custom/data", {})
        return f'''
        <div style="padding: 20px; background: #f0f0f0; border-radius: 8px;">
            <h3>Custom Widget</h3>
            <p>Value: {custom.get("value", "N/A")}</p>
            <p>{custom.get("message", "Loading...")}</p>
            <small>Server: {server_type}</small>
        </div>
        '''

# Display it
custom = CustomWidget()
custom
```

## Summary

This demo shows:

1. **Development mode**: Just `start_infrastructure(mode="development")` and go!
2. **Custom widgets**: Easy to create with `APIDisplay` and `register_endpoint`
3. **Production mode**: Run `enable_production_mode()` when ready for SyftBox
4. **No file pollution**: Your package stays clean in development mode

The key insight is that syft-widget is now a true library that you import and use, not something you fork or modify!
