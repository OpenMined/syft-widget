# Interactive Widget Example

This example shows a widget with a button that:
- Is disabled (grayed out) when the server is not available
- Becomes clickable when the server goes online
- Can perform actions that communicate with the server

```python
from syft_widget import InteractiveWidget, create_server, run_server_in_thread
import time
```

```python
# Create server instance
server = create_server()

# Create interactive widget
widget = InteractiveWidget(server_instance=server)
print("Widget created at:", time.strftime("%Y-%m-%d %H:%M:%S"))
print("The button will be disabled until the server starts...")
widget.display()
```

```python
# Start the server after 5 seconds
print("Starting server in 5 seconds...")
print("Watch the button become enabled when the server goes online!")
run_server_in_thread(delay=5)
print("\nThe button will automatically enable when the server is detected.")
print("You can also click 'Refresh Status' to manually check.")
```

```python
import syft_core as sc
c = sc.Client.load()
c.sync_folder.parent # path to SyftBox folder. apps go in SyftBox/apps
```

```python
# Clean up
widget.stop()
```
