# Managed Widget Example

This demonstrates the three-stage widget lifecycle:

1. **Checkpoint** (🔴): Shows snapshot data immediately
2. **Thread Server** (🟡): Automatically starts a Python thread server
3. **SyftBox App** (🟢): Transitions to SyftBox app when available

The widget will:
- Check if `syft-widget` exists in SyftBox/apps
- Clone it from GitHub if it doesn't exist
- Wait for SyftBox to launch the app
- Automatically transition to the SyftBox server when ready

```python
from syft_widget import ManagedTimeWidget
import time
```

```python
# Create a managed widget
# This will automatically handle all three stages
widget = ManagedTimeWidget(
    app_name="syft-widget",
    repo_url="https://github.com/OpenMined/syft-widget",
    thread_server_port=8001
)

print("Widget created at:", time.strftime("%Y-%m-%d %H:%M:%S"))
print("Watch the status transition through the stages...")
widget.display()
```

## What's happening:

1. The widget immediately shows checkpoint data (frozen timestamp)
2. A Python thread server starts automatically on port 8001
3. In the background, it checks for the SyftBox app:
   - If not found: clones from GitHub
   - If found: waits for SyftBox to launch it
4. Once the SyftBox app is running, it switches to that server
5. The Python thread server is then stopped

The status indicator shows the current stage:
- 🔴 Checkpoint data
- 🟡 Python thread server  
- 🟢 SyftBox app server

```python
# Check the current stage
print(f"Current stage: {widget.current_stage}")
print(f"Thread server running: {widget.thread_server is not None}")
print(f"SyftBox monitoring: {widget.syftbox_manager is not None}")
```

```python
# Clean up when done
widget.stop()
```
