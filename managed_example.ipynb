1. **Checkpoint** (🔴): Shows snapshot data immediately

The widget will:
- Check if `syft-widget` exists in SyftBox/apps
- Clone it from GitHub if it doesn't exist
- Wait for SyftBox to launch the app

## What's happening:

1. The widget immediately shows checkpoint data (frozen timestamp)
2. A Python thread server starts automatically on port 8001
3. In the background, it checks for the SyftBox app:
   - If not found: clones from GitHub
   - If found: waits for SyftBox to launch it
4. Once the SyftBox app is running, it switches to that server
5. The Python thread server is then stopped

The status indicator shows the current stage:
- 🔴 Checkpoint data
- 🟡 Python thread server  
- 🟢 SyftBox app server

```python
from syft_widget import TimeWidget, ManagedTimeWidget, run_server_in_thread
import time
```

```python
# Step 3: Now create a ManagedWidget that handles all three stages
# This will check for SyftBox app and transition automatically

print("\nStep 3: Creating ManagedWidget with SyftBox integration...")
print("This widget will:")
print("  1. Start with checkpoint data")
print("  2. Use its own Python thread server") 
print("  3. Check for syft-widget in SyftBox/apps")
print("  4. Clone from GitHub if needed")
print("  5. Transition to SyftBox server when ready\n")

managed_widget = ManagedTimeWidget(
    app_name="syft-widget",
    repo_url="https://github.com/OpenMined/syft-widget",
    thread_server_port=8002  # Different port to avoid conflict
)

print("Watch the status indicator change:")
print("  🔴 Checkpoint → 🟡 Thread Server → 🟢 SyftBox App")
print(f"\nNote: SyftBox app may be running on port 57016 or similar")
managed_widget.display()
```

## Summary

We've seen three stages:

1. **Basic Widget**: Shows checkpoint data, then updates when we manually start a server
2. **ManagedWidget**: Automatically handles all transitions:
   - Starts with checkpoint data
   - Launches its own Python thread server
   - Monitors for SyftBox app availability
   - Transitions seamlessly between stages

The ManagedWidget provides a complete solution for widgets that need to work both in development (with thread servers) and production (with SyftBox apps).
